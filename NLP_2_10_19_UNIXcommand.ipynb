### 11. タブをスペースに置換
##### タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ

```python
result='./data/2/2-11-python.txt'
with open(result, mode='a') as fw:
    
    with open(datafilename) as fr:
        for line in fr:
            #print(line.replace('\t', ' '), end='')
            fw.write(line.replace('\t', ' '))
            
```

### 16. ファイルをN分割する
##### 自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．．

#### ■Python

```python
n=100

with open(datafilename, 'r') as file:
    lines=file.readlines()

#1ファイルの行数
length=len(lines)//n

for i in range(n):
    resultfilename='./data/2/16-python-'+str(i)+'.txt'
    with open(resultfilename,'w') as fw:
        fw.write(str(lines[i*length:(i+1)*length]))
```

#### ■UNIXコマンド

```python
!split -l 10 ./data/2/popular-names.txt ./data/2/2-16-unix10-
```

### 19. 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
##### 各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

#### ■Python

```python
import collections

l_1d = df[0].values.tolist()
c = collections.Counter(l_1d)
for k, v in sorted(c.items(), key=lambda x: -x[1]):
    print(str(k) + ": " + str(v))
```

#### ■UNIXコマンド

```python
!cut --fields=1 ./data/2/popular-names.txt | sort | uniq -c | sort -nr
```
